```python
Retina_images_dict={
    "DR":list(directory1.glob('DR/*.png')),
    "NO_DR":list(directory1.glob('NO_DR/*.jpg'))
}
```

```python
Retina_images_dict
```

```python
keys = ['DR','NO_DR']
```

```python
resized,labels=[],[]
for Retina_name,images in Retina_images_dict.items():
    for image in images:
        print("hi")
        img=cv2.imread(str(image))
        resized_image=cv2.resize(img,(224,224))
        resized.append(resized_image)
        labels.append(keys.index(Retina_name))
Retina_images_dict
```

```python
def print_image(i,j):
    plt.imshow(i)
    plt.title(keys[j])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(resized,labels,test_size=0.25,random_state=0)
```

```python
print_image(x_train[15],y_train[15])
```

```python
print_image(x_test[23],y_test[23])
```

```python
x_train_scaled=np.array(x_train)/224
x_test_scaled=np.array(x_test)/224
```

```python
y_train=np.array(y_train)
y_train.shape
```

```python
x_train_scaled.shape
```

```python
from keras.layers import Dense,Dropout,Flatten
from tensorflow.keras.models import *
from keras.applications.mobilenet import MobileNet,preprocess_input
model = MobileNet(weights="imagenet")
for i in model.layers:
    i.trainable = False
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Flatten,Dense
from tensorflow.keras.models import Sequential
base_model = MobileNetV2(weights='imagenet',include_top=False,input_shape=(224,224,3))
model1 = Sequential()
model1.add(base_model)
model1.add(Flatten())
model1.add(Dense(1024,activation="relu"))
model1.add(Dense(512,activation="relu"))
model1.add(Dense(2,activation="softmax"))
model1.compile(optimizer='adam',loss = 'sparse_categorical_crossentropy',metrics=['accuracy'])
model1.summary()
```

```python
model1.fit(x_train_scaled,np.array(y_train),epochs=1)
```

```python
model1.evaluate(x_test_scaled,np.array(y_test))
```

```python
model1.save('RetinaNet.h5')
```

```python
from tensorflow.keras.models import load_model
model = load_model('RetinaNet.h5')
def preprocess_image(image_path):
    from tensorflow.keras.preprocessing import image
    img = image.load_img(image_path, target_size=(224, 224))
    img = image.img_to_array(img)
    img = img / 255.0 
    img = np.expand_dims(img, axis=0) 
    return img
def predict_image(image_path):
    img = preprocess_image(image_path)
    prediction = model.predict(img)

    if (prediction > 0.5).any():
        return 'DR'
    else:
        return 'NO_DR'

image_path = 'Eye_disease/val/No_DR/0a74c92e287c.png'
result = predict_image(image_path)
print(f'The image is a {result}')
image_path
```
